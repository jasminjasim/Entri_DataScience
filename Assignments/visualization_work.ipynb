```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
import seaborn as sns
```

```python
titanic=sns.load_dataset("titanic")
titanic=pd.DataFrame(titanic)
titanic
```

```python
# 1. Countplot: Passenger Count by Embarked Port
# ● Create a countplot to display the number of passengers who boarded from each
# port (embarked), separated by sex.
titanic.columns
titanic['embarked'].value_counts()
sns.countplot(data=titanic, x='embarked', hue='sex')
plt.title('Passenger Count by Embarked Port') 
plt.xlabel('Embarked Port')
plt.ylabel('Count')   
plt.show()   
```

Only almost 50% of embarked port 'S' had females 

```python
# 2. Swarmplot: Age Distribution across Classes
# ● Create a swarmplot to visualize the distribution of age across passenger classes
# (pclass), categorized by sex.
sns.swarmplot(data=titanic, x='class', y='age', hue='sex')
plt.title('Age Distribution across Classes by Sex')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
# To get the mean age for each passenger class:
titanic.groupby('class')['age'].mean()
```

From the swarm plot we got an idea that in third class the majority are youngsters.ie Youth are more entered in third class

```python
# 3. Pie Chart: Gender Distribution
# ● Create a pie chart using Matplotlib to show the percentage of male vs. female
# passengers. - Add percentage labels using autopct - Use explode for one slice to
# highlight it.

gender_count=titanic['sex'].value_counts()
labels=gender_count.index
values=gender_count.values
explode=(0.1,0 )
plt.figure(figsize=(6,6))
plt.pie(values,labels=labels,autopct='%1.1f%%',startangle=90,explode=explode,colors=['lightblue','pink'])
plt.title('Distirubution of Titanic Passengers by sex')
plt.show()
```

```python
# 4. Heatmap: Correlation Matrix
# ● Create a heatmap to show correlations between age, fare, pclass, sibsp, parch,
# and survived. (Use seaborn.heatmap with annotations)
corr_matrix=['age','fare','pclass','sibsp','parch','survived']
corr=titanic[corr_matrix].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix of Titanic Dataset')
plt.show()
```

```python

# 5. Violin Plot: Fare by Class and Gender
# ● Create a violin plot to visualize fare distribution across classes, separated by
# gender.
sns.violinplot(data=titanic,x='class',y='fare',hue='sex',split=True)
plt.title('Fare Distribution by Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

Fare is high for the first class and it was utilized more by male

```python

# 6. Scatter Plot: Age vs Fare
# ● Create a scatter plot to explore the relationship between age and fare, colored by
# survival status.
# ● Use a seaborn scatterplot with hue='survived'. Each plot must have titles, labels,
# and legends where needed.

sns.scatterplot(data=titanic, x='age', y='fare', hue='survived', palette={0: 'red', 1: 'green'})
plt.title('Age vs Fare by Survival Status')
plt.xlabel('Age')   
plt.ylabel('Fare')
plt.legend(title='Survived') 
plt.show()  
```
