```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns   
```

```python
df = pd.read_csv(r'C:\Users\jasmi\Downloads\ML Classes\Class Works\bigmart_data-3.xls')
df=df.dropna()
df.head()



```

```python
df.isnull().sum()
```

```python
df.groupby(['Item_Type'])['Item_MRP'].mean().plot(kind='line',figsize=(12,6))
data=df.groupby(['Item_Type'])['Item_MRP'].mean()
plt.figure(figsize=(12,6))
plt.plot(data,color='red',marker='o',linestyle='dashed',linewidth=2,markersize=12)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Item_Weight'],df['Item_MRP'],'o',color='red')
plt.xlabel('Item Weight')       
plt.ylabel('Item MRP')
plt.title('Item Weight vs Item MRP')
plt.show()
```

```python
#what is the mean sales of each outlet type
data1=df.groupby(['Outlet_Type'])['Item_Outlet_Sales'].mean()#=.plot(kind='bar',figsize=(12,6),color='orange')
plt.xlabel('Outlet Type')       
plt.ylabel('Item Outlet Sales')
plt.bar(data1.index,data1,color='orange')
plt.title('Outlet Type vs Item Outlet Sales')   
```

```python
plt.pie(data1,labels=data1.index,autopct='%1.f%%')
plt.title('Outlet Type vs Item Outlet Sales')   
```
